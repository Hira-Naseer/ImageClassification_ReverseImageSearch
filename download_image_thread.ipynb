```python

def get_images_list(query, per_page):
    folder = path + '/' + query
    if not os.path.exists(folder):
      os.makedirs(folder)

    width = 256
    height = 256
    num_images = 80
    API_KEY = 'Wn8y02D0ugJjMBAynIYlquLlD2N1r21qJPCtrK6jMWR8n0HNTwKtF39h'
    url = 'https://api.pexels.com/v1/search'
    query_params = {
        'query': query,
        'per_page': num_images,
        'page': per_page,
        'size': f'{width}x{height}'
    }
    headers = {'Authorization': API_KEY}
    response = requests.get(url, headers=headers, params=query_params)

    if response.status_code == 200:
        data = response.json()

        photos = data['photos']
        images_urls = [photo['src']['original'] for photo in photos]

        return images_urls
    else:
        print('Error in get_images_list()')
        return []

def save_image(url, folder_name,thread):
    photo_response = requests.get(url)
    if photo_response.status_code == 200:
        filename = url.split('-')[-1]
        filename = f'{path}/{folder_name}/{filename}'

        with open(filename, 'wb') as f:
            f.write(photo_response.content)
            print(f'Saved {filename} successfully -- Thread/Page {thread}')
            return True
    else:
        print('Error in save_image()')
        return False


def save_list_images(images_urls, folder_name, thread):
    for url in images_urls:
        save_image(url, folder_name, thread)


```

```python
def run(query):
  threading_list = []
  for i in range(1, 35):
      images_urls = get_images_list(query, i)
      print(f'Page {i} has {len(images_urls)} images, Adding thread')

      t = threading.Thread(target=save_list_images, args=(images_urls, query, i))
      threading_list.append(t)

  for i,t in enumerate(threading_list):
      print(f'Starting thread, {i+2}')
      t.start()

  for i,t in enumerate(threading_list):
      print(f'Joining thread, {i+2}')
      t.join()

  print('Done')


```

```python
run("ship")
```
